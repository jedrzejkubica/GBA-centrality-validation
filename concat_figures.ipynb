```python
import matplotlib.pyplot
import matplotlib.image
```

```python
paths = [
    "figures/MMAF_CDF_GBA_vs_RWR.png",
    "figures/NOA_CDF_GBA_vs_RWR.png",
    "figures/HYPCARD_CDF_GBA_vs_RWR.png",
    "figures/DILCARD_CDF_GBA_vs_RWR.png",
    "figures/BC_CDF_GBA_vs_RWR.png",
    "figures/CC_CDF_GBA_vs_RWR.png"
]

labels = ["(a)", "(b)", "(c)", "(d)", "(e)", "(f)"]
```

```python
fig, axes = matplotlib.pyplot.subplots(2, 3, figsize=(15, 10))

for ax, path, label in zip(axes.ravel(), paths, labels):
    img = matplotlib.image.imread(path)
    ax.imshow(img)
    ax.axis("off")
    
    ax.text(
        0.02, 0.95, label,
        transform=ax.transAxes,
        fontsize=16,
        fontweight="normal",
        va="top", ha="left"
    )

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.savefig("figures/all_CDF_GBA_vs_RWR.png", dpi=300)
matplotlib.pyplot.show()
```
